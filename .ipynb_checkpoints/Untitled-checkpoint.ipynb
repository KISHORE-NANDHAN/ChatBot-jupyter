```python
import os
from langchain_huggingface import HuggingFaceEndpoint
from dotenv import load_dotenv

# Load API key securely
load_dotenv()
sec_key = os.getenv("HUGGINGFACEHUB_API_TOKEN")

# Initialize Hugging Face LLM
repo_id = "mistralai/Mistral-7B-Instruct-v0.2"
llm = HuggingFaceEndpoint(
    repo_id=repo_id,
    task="text-generation",
    max_length=700,
    temperature=0.7,
    huggingfacehub_api_token=sec_key  # Corrected key
)

# Faculty JSON Data
faculty_data = {
    "name": "Dr. D. Veeraiah",
    "designation": "Professor and HOD at Lakireddy Balireddy College of Engineering, Mylavaram",
    "department": "Computer Science and Engineering (CSE)",
    "experience": "16+ Years",
    "email": ["hodcse@Lbrce.ac.in", "veeraiahdvc@gmail.com"],
    "phone": ["+91 8919810599", "+91 8333905834"],
    "education": {
        "phd": "Computer Science and Engineering, JNTU Kakinada, 2017",
        "mtech": "Computer Science and Engineering, JNTU Hyderabad, 2009",
        "btech": "Computer Science and IT, JNTU Hyderabad, 2005"
    },
    "research_areas": ["Machine Learning", "Network Security", "Internet of Things"],
    "publications": {"journals": 27, "conferences": 15, "patents": 5},
    "professional_memberships": ["CSI", "ISTE", "IAENG", "IEEE"],
    "achievements_and_awards": [
        "Research Excellence Award - 2023",
        "Academic Titan Achievement Award - 2023",
        "Award of Excellence in Research - 2022",
        "Best Teacher Award by SOLETE - 2021"
    ]
}

# Convert JSON to a Structured Prompt
structured_text = f"""
Convert the following structured faculty information into a natural human-like response:

Name: {faculty_data['name']}
Designation: {faculty_data['designation']}
Department: {faculty_data['department']}
Experience: {faculty_data['experience']}
Email: {', '.join(faculty_data['email'])}
Phone: {', '.join(faculty_data['phone'])}
Education:
  - PhD: {faculty_data['education']['phd']}
  - MTech: {faculty_data['education']['mtech']}
  - BTech: {faculty_data['education']['btech']}
Research Areas: {', '.join(faculty_data['research_areas'])}
Publications:
  - Journals: {faculty_data['publications']['journals']}
  - Conferences: {faculty_data['publications']['conferences']}
  - Patents: {faculty_data['publications']['patents']}
Professional Memberships: {', '.join(faculty_data['professional_memberships'])}
Achievements and Awards: {', '.join(faculty_data['achievements_and_awards'])}

Generate a detailed yet natural human response based on this data in a div tag formatting the information.
"""

# Invoke LLM
response = llm.invoke(structured_text)
print(response)
```
